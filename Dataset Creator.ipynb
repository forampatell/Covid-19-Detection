```python
import pandas as pd
import os
import shutil
```

```python
FILE_PATH="C:\\Users\\Dell\\Desktop\\projects eg\\covid\\covidxray\\metadata.csv"
IMAGES_PATH="C:\\Users\\Dell\\Desktop\\projects eg\\covid\\covidxray\\images"
```

```python
df=pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
TARGET_DIR="C:\\Users\\Dell\\Desktop\\projects eg\\covid\\Dataset\\covid19"
```

```python
count=0
for(i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename=row["filename"]
        image_path=os.path.join(IMAGES_PATH,filename)
        image_copy_path=os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
#         print("MOVING IMAGEs",count)
        count+=1
```
